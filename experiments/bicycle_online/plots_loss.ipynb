```python
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt

experiment_name = "M1000_b500_s2_dqn6000_pbo3000_25_idqn2500_40"
show_dqn = True; show_pbo_linear = False; show_pbo_deep = True; show_idqn = True
max_bellman_iterations = 8
seed = 7
```

```python
if show_dqn:
    dqn_loss = np.load(f"figures/{experiment_name}/DQN/{max_bellman_iterations}_L_{seed}.npy")

    for bellman_iteration, loss_fit in enumerate(dqn_loss):
        plt.plot(loss_fit, label=f"iteration {bellman_iteration}")

    plt.legend()
    _ = plt.title("DQN training loss")
```

```python
if show_pbo_linear:
    pbo_linear_loss = np.load(f"figures/{experiment_name}/PBO_linear/{max_bellman_iterations}_L_{seed}.npy")
    plt.plot(pbo_linear_loss.flatten())
    _ = plt.title(f"PBO linear training loss with {pbo_linear_loss.shape[1]} fitting steps")
```

```python
if show_pbo_deep:
    pbo_deep_loss = np.load(f"figures/{experiment_name}/PBO_deep/{max_bellman_iterations}_L_{seed}.npy")
    plt.plot(pbo_deep_loss.flatten()[:])
    _ = plt.title(f"PBO deep training loss with {pbo_deep_loss.shape[1]} fitting steps")
```

```python
if show_idqn:
    idqn_loss = np.load(f"figures/{experiment_name}/IDQN/{max_bellman_iterations}_L_{seed}.npy")
    plt.plot(idqn_loss.flatten())
    _ = plt.title(f"IDQN training loss with {idqn_loss.shape[1]} fitting steps")
```
