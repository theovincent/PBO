```python
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt

max_bellman_iterations = 9
seed = 8
```

```python
fqi_loss = np.load(f"figures/data/FQI/{max_bellman_iterations}_L_{seed}.npy")

for bellman_iteration, loss_fit in enumerate(fqi_loss):
    plt.plot(loss_fit, label=f"iteration {bellman_iteration}")

plt.legend()
_ = plt.title("FQI training loss")
```

```python
pbo_linear_loss = np.load(f"figures/data/PBO_linear/{max_bellman_iterations}_L_{seed}.npy")

plt.plot(pbo_linear_loss.flatten())

_ = plt.title(f"PBO linear training loss with {pbo_linear_loss.shape[1]} fitting steps")
```

```python
pbo_deep_loss = np.load(f"figures/data/PBO_deep/{max_bellman_iterations}_L_{seed}.npy")

plt.plot(pbo_deep_loss.flatten())

_ = plt.title(f"PBO deep training loss with {pbo_deep_loss.shape[1]} fitting steps")
```
