```python
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt

experiment_name = "h1000_q60_10_n500_N15000_b500_s2_f8000_lr4_pbo2_2_2_2_f4000_25_lr5_e2_w30_30"
show_dqn = True; show_pbo_linear = False; show_pbo_deep = True; show_idqn = False
max_bellman_iterations = 15
seed = 7
```

```python
if show_dqn:
    fqi_loss = np.load(f"figures/{experiment_name}/DQN/{max_bellman_iterations}_L_{seed}.npy")

    for bellman_iteration, loss_fit in enumerate(fqi_loss):
        plt.plot(loss_fit, label=f"iteration {bellman_iteration}")

    plt.legend()
    _ = plt.title("DQN training loss")
```

```python
if show_pbo_linear:
    pbo_linear_loss = np.load(f"figures/{experiment_name}/PBO_linear/{max_bellman_iterations}_L_{seed}.npy")
    plt.plot(pbo_linear_loss.flatten()[:])
    _ = plt.title("PBO linear training loss")
```

```python
if show_pbo_deep:
    pbo_deep_loss = np.load(f"figures/{experiment_name}/PBO_deep/{max_bellman_iterations}_L_{seed}.npy")
    plt.plot(pbo_deep_loss.flatten()[:])
    _ = plt.title("PBO deep training loss")
```

```python
if show_idqn:
    ifqi_loss = np.load(f"figures/{experiment_name}/IDQN/{max_bellman_iterations}_L_{seed}.npy")
    plt.plot(ifqi_loss.flatten())
    _ = plt.title(f"IDQN training loss with {ifqi_loss.shape[1]} fitting steps")
```
