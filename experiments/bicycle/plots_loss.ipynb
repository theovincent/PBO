```python
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt

experiment_name = "b8_q30_s70000_b1000_pbo_500_15"
show_fqi = True; show_pbo_linear = True; show_pbo_deep = True; show_ifqi = True
max_bellman_iterations = 8
seed = 4
```

```python
if show_fqi:
    fqi_loss = np.load(f"figures/{experiment_name}/FQI/{max_bellman_iterations}_L_{seed}.npy")

    for bellman_iteration, loss_fit in enumerate(fqi_loss):
        plt.plot(loss_fit, label=f"iteration {bellman_iteration}")

    plt.legend()
    _ = plt.title("FQI training loss")
```

```python
if show_pbo_linear:
    pbo_linear_loss = np.load(f"figures/{experiment_name}/PBO_linear/{max_bellman_iterations}_L_{seed}.npy")
    plt.plot(pbo_linear_loss.flatten())
    _ = plt.title(f"PBO linear training loss with {pbo_linear_loss.shape[1]} fitting steps")
```

```python
if show_pbo_deep:
    pbo_deep_loss = np.load(f"figures/{experiment_name}/PBO_deep/{max_bellman_iterations}_L_{seed}.npy")
    plt.plot(pbo_deep_loss.flatten()[:])
    _ = plt.title(f"PBO deep training loss with {pbo_deep_loss.shape[1]} fitting steps")
```

```python
if show_ifqi:
    ifqi_loss = np.load(f"figures/{experiment_name}/IFQI/{max_bellman_iterations}_L_{seed}.npy")
    plt.plot(ifqi_loss.flatten())
    _ = plt.title(f"IFQI training loss with {ifqi_loss.shape[1]} fitting steps")
```
